```python
%pip install pandas matplotlib numpy ipykernel
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("C:/Users/subuh/PycharmProjects/PythonProject1/my_clean_dataset.csv")
df.head()
```

1.Top Selling Products

```python

top_products=df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
top_products.head(10)
plt.Figure(figsize=(10,6))
top_products.plot(kind='bar')
plt.title('top_selling_products')
plt.xlabel('product_name')
plt.ylabel('sales')
plt.show() 
```

KEY INSIGHT: It shows the products with highest sales.

2.Customer Segmentation


```python
Top_cus_segment=df.groupby('segment')['sales'].sum()
Top_cus_segment.head()
Top_cus_segment.plot(kind='bar')
plt.title('customer segmentation')
plt.xlabel('segment')
plt.ylabel('sales')
plt.show() 
```

KEY INSIGHT: It identifies that which customer segment spend more. Consumer segment generates the highest sales.

3.Monthly Sales Trend

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['month']=df['order_date'].dt.month
monthly_trend= df.groupby('month')['sales'].sum()
monthly_trend
monthly_trend.plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('sales')
plt.show() 
```

KEY INSIGHT:This shows how sales change over time and helps identify seasonal trends. 
Sales are higher in last two months(November & December) and lowest in first two months.

4.Outliers

```python
plt.scatter(df['sales'], df['profit'])
plt.xlabel('sales')
plt.ylabel('Profit')
plt.title('Outliers')
plt.show()
```

KEY INSIGHT:Most orders have positive relationship b/w sales and profit but some show unusual pattern with high sales and loss. It happens usually due to heavy discounts.
